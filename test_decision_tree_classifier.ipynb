```python
import pandas as pd

fruits_file_path = 'frutas.csv'
fruits_data = pd.read_csv(fruits_file_path)

fruits_data = fruits_data.sample(frac=1).reset_index(drop=True)

print(fruits_data.head())

fruits_data.columns
```

```python
y = fruits_data.classe
print(y)
```

```python
fruits_features = ['area', 'perimetro', 'eixo_maior', 'eixo_menor', 'excentricidade',
    'eqdiasq', 'solidez', 'area_convexa', 'extensao', 'proporcao',
    'redondidade', 'compactidade', 'fator_forma_1', 'fator_forma_2',
    'fator_forma_3', 'fator_forma_4', 'RR_media', 'RG_media', 'RB_media',
    'RR_dev', 'RG_dev', 'RB_dev', 'RR_inclinacao', 'RG_inclinacao',
    'RB_inclinacao', 'RR_curtose', 'RG_curtose', 'RB_curtose',
    'RR_entropia', 'RG_entropia', 'RB_entropia', 'RR_all', 'RG_all',
    'RB_all']

X = fruits_data[fruits_features]

X.describe()
X.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

fruits_random_forest_model = RandomForestClassifier(random_state=1)
fruits_random_forest_model.fit(X, y)
fruits_random_forest_model.predict(X)
```

```python
from sklearn.tree import DecisionTreeClassifier

fruits_decision_tree_model = DecisionTreeClassifier(random_state=1)
fruits_decision_tree_model.fit(X, y)
fruits_decision_tree_model.predict(X)
```
